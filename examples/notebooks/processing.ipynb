```python
import sys
sys.path.append("../../")

import radvis as rv
import matplotlib.pyplot as plt
%matplotlib widget
```

```python
ATLAS_IMAGE = "../images/mni_icbm152_nlin_sym_09a/mni_icbm152_t1_tal_nlin_sym_09a.nii"
EXAMPLE_IMAGE = "../images/sub-A00028185_ses-NFB3_T1w.nii.gz"
EXAMPLE_MASK = "../images/sub-A00028185_ses-NFB3_T1w_brainmask.nii.gz"

atlas = rv.load_image(ATLAS_IMAGE)
img = rv.load_image(EXAMPLE_IMAGE)
mask = rv.load_image(EXAMPLE_MASK)
```

```python
img = rv.percentile_clipping(img, lower_percentile=5, upper_percentile=95)
img = rv.process.intensity_normalization(img)
denoised_img = rv.process.noise_reduction(img, sigma=0.5)
```

```python
img_slicer = rv.RadSlicer(img, 2)
denoised_img_slicer = rv.RadSlicer(denoised_img, 2)
```

```python
img_slicer.display()
denoised_img_slicer.display()
```
