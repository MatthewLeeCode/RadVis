```python
import sys
sys.path.append("../../")

import radvis as rv
import numpy as np
import matplotlib.pyplot as plt
%matplotlib widget
```

```python
IMAGE_PATH = "../images/sub-A00028185_ses-NFB3_T1w.nii.gz"

# Load the RadImage
image = rv.load_image(IMAGE_PATH)
```

```python
slicer = rv.RadSlicer(image, 2, show_slider=False)

red_mask = np.zeros_like(image.image_data) 
red_mask[5:100, 5:100, 5:100] = 1

blue_mask = np.zeros_like(image.image_data)
blue_mask[70:150, 70:150, 70:150] = 1

slicer.add_mask(red_mask, color="red")
slicer.add_mask(blue_mask, color="blue")

slicer.display(initial_index=50)
```
